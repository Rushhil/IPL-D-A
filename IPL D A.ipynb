```python
iplmatch_data['Season'] = pd.DatetimeIndex(iplmatch_data['date']).year
iplmatch_data.head()
```

```python
match_per_season = iplmatch_data.groupby(['season'])['match_id'].count().reset_index().rename(columns={'match_id':'matches'})
match_per_season
```

```python
sns.countplot(iplmatch_data['Season'],saturation=0)
plt.xticks(rotation=90, fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('\nSeason--->', fontsize=15)
plt.ylabel('\nCount----> ', fontsize=15)
plt.title('Total matches played in each season', fontsize = 19, fontweight = "bold")
```

```python
season_data=iplmatch_data[['match_id','Season']].merge(iplball_data, left_on = 'match_id', right_on = 'match_id', how = 'left').drop('match_id', axis = 1)
season_data.head()
```

```python
toss=iplmatch_data['toss_winner'].value_counts()
ax = plt.axes()
ax.set(facecolor = "cyan")
sns.set(rc={'figure.figsize':(25,10)},style='darkgrid')
ax.set_title('No. of tosses won by each team',fontsize=19,fontweight="bold")
sns.barplot(y=toss.index, x=toss, orient='h',palette="icefire",saturation=0)
plt.xlabel('# of tosses won')
plt.ylabel('Teams')
plt.show()
```

```python
ax = plt.axes()
ax.set(facecolor = "cyan")
sns.countplot(x='season', hue='toss_decision', data=iplmatch_data,palette="magma",saturation=0)
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=15)
plt.xlabel('\n Season---->.',fontsize=15)
plt.ylabel('Count---->',fontsize=15)
plt.title('Toss decision across seasons',fontsize=19,fontweight="bold")
plt.show()
```

```python
player = (iplball_data['striker']=='S Dhawan')
df_dhawan=iplball_data[player]
df_dhawan.head()
```

```python
def count(df_dhawan,runs):
    return len(df_dhawan[df_dhawan['runs_off_bat']==runs])*runs
```

```python
print("Runs scored from 1's :",count(df_dhawan,1))
print("Runs scored from 2's :",count(df_dhawan,2))
print("Runs scored from 3's :",count(df_dhawan,3))
print("Runs scored from 4's :",count(df_dhawan,4))
print("Runs scored from 6's :",count(df_dhawan,6))
```

```python
df_dhawan['wicket_type'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,rotatelabels=True)
plt.title("Dismissal of S Dhawan In Differnt Manners",fontweight="bold",fontsize=15)
plt.show()
```

```python
runs = iplball_data.groupby(['striker'])['runs_off_bat'].sum().reset_index()
runs.columns = ['striker', 'runs_off_bat']
y = runs.sort_values(by='runs_off_bat', ascending = False).head(10).reset_index().drop('index', axis=1)
y
```

```python
ax = plt.axes()
ax.set(facecolor = "cyan")
sns.barplot(x=y['striker'],y=y['runs_off_bat'],palette='rocket',saturation=0)
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('\n Players---->',fontsize=15)
plt.ylabel('Total Runs---->',fontsize=15)
plt.title('Top 10 run scorers in IPL',fontsize=19,fontweight="bold")
```
